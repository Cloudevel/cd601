# ```Cloud SQL```.

* ```PostrgeSQL``` versiones 13, 12, 11, 10, 9.6
* ```MySQL``` versiones: 8.0, 5.7, 5.6
* ```MS SQL Server``` versiones: 2019, 2017

````bash
gcloud sql instances create --help
````

## Respaldos.

### Creación de respaldos.

```
gcloud sql backups create  --instance <instancia>
```

### Calendarizacion de respaldos.

```
gcloud sql instances patch <instancia> –backup-start-time <HH:MM>
```
